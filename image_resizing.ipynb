```python
import cv2
import os
import math

# Girdi ve çıktı klasörlerini belirleyin
input_folder = "Tabak/"
output_folder = "Tabak_resized"

# Çıktı klasörünü oluştur (yoksa)
os.makedirs(output_folder, exist_ok=True)

# Tüm görüntüler üzerinde döngü
for filename in os.listdir(input_folder):
    if filename.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp')):
        # Görüntüyü yükle
        img_path = os.path.join(input_folder, filename)
        img = cv2.imread(img_path)
        
        if img is None:
            print(f"Hata: {filename} okunamadı, atlanıyor...")
            continue
        
        # Mevcut boyutlar
        original_height, original_width = img.shape[:2]
        original_area = original_width * original_height
        
        # Zaten 2MP'den küçükse atla (opsiyonel)
        if original_area <= 2000000:
            print(f"{filename} zaten 2MP'den küçük, atlanıyor...")
            continue
        
        # En boy oranını koruyarak yeni boyutları hesapla
        target_area = 2000000
        ratio = original_width / original_height
        new_width = int(math.sqrt(target_area * ratio))
        new_height = int(math.sqrt(target_area / ratio))
        
        # Yeniden boyutlandır
        resized_img = cv2.resize(
            img, 
            (new_width, new_height), 
            interpolation=cv2.INTER_AREA
        )
        
        # Kaydet
        output_path = os.path.join(output_folder, filename)
        cv2.imwrite(output_path, resized_img)
        print(f"{filename} başarıyla 2MP yapıldı: {new_width}x{new_height}")

print("Tüm işlemler tamamlandı!")
```

```python
import os
import cv2

# AYARLAR (Kendinize Göre Düzenleyin)
input_folder = "Tabak_resized/"  # Orijinal fotoğrafların klasörü
output_folder = "Tabak_images"  # Yeni kaydedilecek klasör
base_name = "Tabak"  # Temel isim formatı
start_number = 1  # Başlangıç numarası
jpg_quality = 95  # 0-100 arası kalite (yüksek değer = daha iyi kalite)

# Klasörü oluştur
os.makedirs(output_folder, exist_ok=True)

# Tüm JPG'leri al ve sırala
all_images = sorted([f for f in os.listdir(input_folder) if f.lower().endswith(".jpg")])

success_count = 0
fail_count = 0

for index, filename in enumerate(all_images, start=start_number):
    # Yeni dosya adı
    new_name = f"{base_name}_{index}.jpg"
    
    # Dosya yolları
    input_path = os.path.join(input_folder, filename)
    output_path = os.path.join(output_folder, new_name)
    
    # Görseli OpenCV ile oku
    img = cv2.imread(input_path)
    
    if img is not None:
        # Yeniden kaydet (BGR formatında kaydeder)
        cv2.imwrite(output_path, img, [cv2.IMWRITE_JPEG_QUALITY, jpg_quality])
        print(f"Başarılı: {filename} -> {new_name}")
        success_count += 1
    else:
        print(f"Hata: {filename} okunamadı!")
        fail_count += 1

print(f"\nİşlem tamamlandı!\nBaşarılı: {success_count}\nBaşarısız: {fail_count}")
```
