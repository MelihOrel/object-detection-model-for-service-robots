```python
from ultralytics import YOLO
import cv2
```

```python
# Eitilmi modeli ykle
model = YOLO("runs/detect/train/weights/best.pt")

# Videoyu a
video_path = "test_video.mp4"
cap = cv2.VideoCapture(video_path)
fps = cap.get(cv2.CAP_PROP_FPS)
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

# kt videosu iin codec ve boyut ayarlar
fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # 'XVID' veya 'MJPG' de tercih edilebilir
output_video = cv2.VideoWriter("output_video.mp4", fourcc, fps, (width, height))

with open("data.txt", "w") as f:
    frame_number = 0
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        # Model ile tahmin yap
        results = model(frame)
        annotated_frame = results[0].plot()  # Bounding box'lar iz

        # ereve boyutunu kontrol et ve videoyla eletir
        if annotated_frame.shape[1] != width or annotated_frame.shape[0] != height:
            annotated_frame = cv2.resize(annotated_frame, (width, height))

        # kt videosuna ekle
        output_video.write(annotated_frame)

        # Zaman bilgisini hesapla
        timestamp = frame_number / fps
        # Tespit edilen nesneleri kontrol et
        for box in results[0].boxes:
            confidence = float(box.conf[0])
            # Sadece gven skoru %80 (0.8) ve zerinde olan tespitler TXT dosyasna kaydedilir
            if confidence < 0.8:
                continue
            class_id = int(box.cls[0])
            x1, y1, x2, y2 = map(int, box.xyxy[0])
            label = model.names[class_id]
            f.write(f"Frame {frame_number}, {timestamp:.2f}s, {label} ({confidence:.2f}) at ({x1},{y1},{x2},{y2})\n")

        # Canl izleme
        cv2.imshow("Canl zleme", annotated_frame)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
        frame_number += 1

# Kaynaklar serbest brak
cap.release()
output_video.release()
cv2.destroyAllWindows()
print("lem tamamland! ktlar: output_video.mp4 ve data.txt")
```
